```python
import pandas as pd

# Define the path to the test data directory
test_dir = r"D:\crack-detection-image-classification-2023\test"

# Get the list of test image file names
test_files = os.listdir(test_dir)

# Create an empty list to store the predictions
predictions = []

# Loop over each test image and make a prediction
for file in test_files:
    img = tf.keras.preprocessing.image.load_img(os.path.join(test_dir, file), target_size=(img_width, img_height))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Create a batch

    # Make the prediction
    prediction = model.predict(img_array)
    predicted_class = 'cracked' if prediction[0][0] < prediction[0][1] else 'uncracked'
    predictions.append([file, predicted_class])

# Save the predictions to a CSV file
df = pd.DataFrame(predictions, columns=['filename', 'class'])
```
